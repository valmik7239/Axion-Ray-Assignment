```python
# plot the graphs by using cleaned dataset
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats
```

```python
# 1.Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(d['KM'], d['TOTALCOST'], alpha=0.5)
plt.title('Mileage (KM) vs Total Cost (TOTALCOST)')
plt.xlabel('Mileage (KM)')
plt.ylabel('Total Cost (TOTALCOST)')
plt.grid(True)
plt.show()
```

```python
# Group by 'STATE' column and count the number of transactions
state_counts = d['STATE'].value_counts()
```

```python
# Plot the bar graph
plt.figure(figsize=(8, 4))
state_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Transactions per State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Transactions', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plotting a line chart - Repair date vs Reporting cost
plt.figure(figsize=(10, 6))
plt.plot(d['REPAIR_DATE'], d['REPORTING_COST'], label='Reporting Cost', color='blue')
plt.title('Repair Date vs Reporting Cost')
plt.xlabel('Repair Date')
plt.ylabel('Reporting Cost')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Box plot for 'REPAIR_AGE' and 'KM'
d[['REPAIR_AGE', 'KM']].boxplot()
plt.title('Box Plot of REPAIR_AGE and KM')
plt.ylabel('Values')
plt.show()
```
